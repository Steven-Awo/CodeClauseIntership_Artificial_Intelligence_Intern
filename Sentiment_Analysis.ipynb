```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer as vad
nltk.download('vader_lexicon')
```

```python
# Read the CSV file into a pandas DataFrame
training_datafile = pd.read_csv('C:/Users/Rotimi O. Awomodu/Documents/Artificial Intelligence Intern(codeclause)/SentimentAnalysis/sentiment-analysis-movie-reviews/train.tsv', sep='\t', nrows=1000)

# training_datafile = pd.read_csv('C:/Users/Rotimi O. Awomodu/Documents/Artificial Intelligence Intern(codeclause)/SentimentAnalysis/SentimentAnalysisDatasete/train.csv', sep='\t', nrows=1000)

# Display the first 100 rows of the DataFrame
training_datafile.head(10)
```

```python
vaderr = training_datafile.copy()
senti = vad()

sennt = ['Positive', 'Neutral', 'Negative', 'Compound']

sentis = [senti.polarity_scores(y) for y in training_datafile['Phrase'].values]

vaderr['Positive Score'] = [y['pos'] for y in sentis]

vaderr['Neutral Score'] = [y['neu'] for y in sentis]

vaderr['Negative Score'] = [y['neg'] for y in sentis]

vaderr['Compound Score'] = [y['neu'] for y in sentis]

scorres = vaderr['Compound Score'].values

tpl = []

for x in scorres:
    if x >= 0.05:
        tpl.append('Positive')
    elif x <= -0.05:
        tpl.append('Negative')
    else:
        tpl.append('Neutral')
        
vaderr['The Total Sentiment'] = tpl

```
